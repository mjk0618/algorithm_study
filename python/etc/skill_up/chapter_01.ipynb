```python
class Base:
    pass

class Sub(Base):
    pass

a = type(Base)
print(a)
b = a[1:2]  # subtype of the original type object 'a'
print(b)
c = b("Sub", (Base,), {})
print(type(c))  # Output: <class 'type'>
print(issubclass(c, Base))  # Output: True
print(b)
print(c)
```

```python
class Base:
    pass

class Sub(Base):
    def __init__(self, x):
        self.x = x

a = type(Base)
b = a[1:2]
c = b("Sub", (Base,), {"__init__": Sub.__init__})

# 아래 두 줄은 같은 결과
# c는 Sub클래스의 서브타입인 새로운 클래스 객체 c를 호출한다는 점이 차이점
# s = Sub(42)
s = c(42)
print(s.x)  # Output: 42
```
