```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

# x, y축의 값의 범위 정하기
plt.xlim([0, 2])
plt.ylim([0, 1])
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

# 제목과 축의 라벨 표시하기
plt.title("y=sin(x) (0 < y < 1)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.ylim([0, 1])

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

# 격자 표시하기
plt.grid(True)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

## 축의 눈금 삽입하기
# plt.xticks(눈금삽입위치, 삽입할 눈금)
positions = np.linspace(0, 2*np.pi, 5)
labels = np.linspace(0, 360, 5)
plt.xticks(positions, labels)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)
positions = np.linspace(0, 2*np.pi, 5)
labels = np.linspace(0, 360, 5)

plt.title("graphs of trigonometric functions")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.xticks(positions, labels)

plt.plot(x, y1, color="k", label="y=sin(x)")
plt.plot(x, y2, color="b", label="y=cos(x)")

plt.legend(["y=sin(x)", "y=cos(x)"])

plt.show()
```
