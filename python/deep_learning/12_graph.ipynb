```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

days = np.arange(1, 11)
weight = np.array([10, 14, 18, 20, 18, 16, 17, 18, 20, 17])

plt.ylim([0, weight.max() + 1])
plt.xlabel("days")
plt.ylabel("weight")

plt.plot(days, weight, "ko-")
plt.show()
```

```python
plt.ylim([0, weight.max() + 1])
plt.xlabel("days")
plt.ylabel("weight")

plt.plot(days, weight, "ko--")
plt.show()
```

```python
x = np.arange(1, 7)
y = [12, 41, 32, 36, 21, 17]

plt.bar(x, y)
plt.show()
```

```python
x = np.arange(1, 7)
y = [12, 41, 32, 36, 21, 17]
labels = ["Apple", "Banana", "Orange", "Pineapple", "Kiwifruit", "Strawberry"]
plt.bar(x, y, tick_label=labels)
plt.show()
```

```python
x = np.arange(1, 7)
y1 = [12, 41, 32, 36, 21, 17]
y2 = [43, 1, 6, 17, 17, 9]
labels = ["Apple", "Banana", "Orange", "Pineapple", "Kiwifruit", "Strawberry"]
plt.bar(x, y1, tick_label=labels)
plt.bar(x, y2, bottom=y1)
plt.legend(("y1", "y2"))
plt.show()
```

```python
data = [60, 20, 10, 5, 3, 2]
plt.pie(data)
plt.axis("equal")
plt.show()
```

```python
data = [60, 20, 10, 5, 3, 2]
labels = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]
plt.pie(data, labels=labels)
plt.axis("equal")
plt.show()
```

```python
data = [60, 20, 10, 5, 3, 2]
labels = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]
explode = [0, 0, 0.1, 0, 0, 0]  # explode를 0~1 사이의 값으로 지정하여 부각

plt.pie(data, labels=labels, explode=explode)
plt.axis("equal")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
x = y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2)/2) / (2*np.pi)

fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1, projection="3d")

xpos = [i for i in range(10)]
ypos = [i for i in range(10)]
zpos = np.zeros(10)

dx = np.ones(10)
dy = np.ones(10)
dz = [i for i in range(10)]

ax.bar3d(xpos, ypos, zpos, dx, dy, dz)
```

```python
np.random.seed(0)

X = np.random.randn(1000)
Y = np.random.randn(1000)
Z = np.random.randn(1000)

fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(1, 1, 1, projection="3d")

x = np.ravel(X)
y = np.ravel(Y)
z = np.ravel(Z)

ax.scatter3D(x, y, z)
plt.show()
```

```python
from matplotlib import cm

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

plt.show()
```
