```python
# 자동으로 데이터타입 변경
# species, island, sex는 텍스트 형식인데 제대로 판단이 안된 것 같음
# 아래 함수를 이용하여 데이터 타입 추론
df = df.convert_dtypes()
df.dtypes
```

```python
# 수동으로 데이터 타입 변경 1
df = df.astype({"species": "category"})

# 수동으로 데이터 타입 변경 2 (한 번에 변경)
df = df.astype({"island": "category", "sex": "category"})

# 수동으로 데이터 타입 변경 3
df["species"] = df["species"].astype("category")
df.dtypes
```

```python
# 기술통계: 수치형 변수와 관련된 기술통계 정보만 제공
df = sns.load_dataset("penguins")
df.describe()
```
