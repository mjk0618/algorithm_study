```python
import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[2, 2] = NA
sample_data_frame.iloc[5:, 3] = NA
```

```python
sample_data_frame
```

```python
## 리스트와이즈 삭제(listwise deletions)
# NaN을 가진 행을 통째로 지우는 것

sample_data_frame.dropna()
```

```python
## 페어와이즈 삭제(pairwise deletions)
# 결손이 적은 열만을 남기는 것
sample_data_frame[[0, 1, 2]]
```

```python
sample_data_frame[[0, 1, 2]].dropna()
```

```python
np.random.seed(0)

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[2, 2] = NA
sample_data_frame.iloc[5:, 3] = NA

sample_data_frame[[0, 2]].dropna()
```

```python
np.random.seed(0)
sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[2, 2] = NA
sample_data_frame.iloc[5:, 3] = NA

sample_data_frame.fillna(0)  # 결측치를 0으로 채움
```

```python
sample_data_frame.fillna(method="ffill")  # NaN 부분을 앞에 있는 데이터르 채움
```

```python
np.random.seed(0)
sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[2, 2] = NA
sample_data_frame.iloc[5:, 3] = NA

sample_data_frame.fillna(sample_data_frame.mean())  # 평균값 대입
```
